```python
#function to remove duplicates
def Remove(duplicate):
    final_list = []
    for num in duplicate:
        if num not in final_list:
            final_list.append(num)
    return final_list
```

```python
def data_clean(twe):
    import nltk
    import csv
    import string
    import re
    from nltk import tokenize
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize
    from nltk.tokenize import TreebankWordTokenizer
    tweet_clean = re.sub(r'https?:\/\/.*[\r\n]*', '', twe)
    tc = re.sub(r'RT', '', tweet_clean)
    article1 = re.sub(r' a ',' ',tc)
    article2 = re.sub(r' an ',' ',article1)
    article3 = re.sub(r' the ',' ',article2)
    sw1 = re.sub(r' has ',' ',article3)
    sw2 = re.sub(r' of ',' ',sw1)
    sw3 = re.sub(r' was ',' ',sw2)
    sw4 = re.sub(r' to ',' ',sw3)
    sw5 = re.sub(r' on ',' ',sw4)
    sw6 = re.sub(r' and ',' ',sw5)
    sw7 = re.sub(r' are ',' ',sw6)
    sw8 = re.sub(r' as ',' ',sw7)
    sw9 = re.sub(r' at ',' ',sw8)
    sw10 = re.sub(r' be ',' ',sw9)
    sw11 = re.sub(r' by ',' ',sw10)
    sw12 = re.sub(r' from ',' ',sw11)
    sw13 = re.sub(r' its ',' ',sw12)
    sw14 = re.sub(r' it ',' ',sw13)
    sw15 = re.sub(r' is ',' ',sw14)
    sw16 = re.sub(r' in ',' ',sw15)
    sw17 = re.sub(r' he ',' ',sw16)
    sw18 = re.sub(r' that ',' ',sw17)
    sw19 = re.sub(r' were ',' ',sw18)
    sw20 = re.sub(r' will ',' ',sw19)
    sw21 = re.sub(r' with ',' ',sw20)
    tokens = nltk.sent_tokenize(sw21)
            #remove punctuations
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]
            #remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_sentence = [w for w in stripped if not w in stop_words]
    filtered_sentence = []
    for w in stripped:
        if w not in stop_words:
            filtered_sentence.append(w)
    WNlemma = nltk.WordNetLemmatizer()
       #Perform Lemmatisation
    lemma = [WNlemma.lemmatize(t) for t in filtered_sentence]
    fs = ' '.join(lemma)
            #row["text"] = fs
       #write dynamically into the file
            #csvwriter.writerow(row)
    return fs
```
